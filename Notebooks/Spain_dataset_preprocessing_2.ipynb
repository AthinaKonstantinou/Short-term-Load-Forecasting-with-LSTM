```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
from math import sqrt
import math
```

```python
#create time features and load features
#loading data fron csv files
My_Data = pd.read_csv("Final_data_1.csv") 
My_Data['time'] = pd.to_datetime(My_Data['time'], utc = True)
My_Data.head()
```

```python
#create column day of the year and day of month 
My_Data['day_of_year'] = My_Data['time'].dt.dayofyear
My_Data['day_of_month'] = My_Data['time'].dt.day
```

```python
#create a column of weekdays (0) and weekend (1) indicator
weekday_weekend = []
for i in range(0,len(My_Data)):
    if My_Data['weekday'][i] > 4.0 :
        weekday_weekend.append(1)
    else: 
        weekday_weekend.append(0)
My_Data['weekdays_weekends'] = weekday_weekend        
```

```python
#find the load value, one, two, three, and four weeks before, create a column in dataframe, 
for j in range(1,5):
    
    load_trend = []
    for i in range(0,len(My_Data)):
        if i < 24*7*j :
            load_trend.append(0)
        else: 
            load_trend.append(My_Data.loc[i-24*7*j, 'total load actual'])
    My_Data[str(j)+'prev_week_load'] = load_trend
```

```python
#Final_Data = My_Data[7*24*4:]
Final_Data = My_Data[7*24*4:].copy(deep=True)
```

```python
Final_Data = Final_Data.reset_index(drop=True)
Final_Data.head()
```

```python
Final_Data.columns
```

```python
#display weather features as sine and cosine
def sine_cosine( dataframe, column ):
    count_values = dataframe[column].nunique()
    pie_degree = 360/count_values
    pie_size = (3.1415*pie_degree)/180
    sine = []
    cosine = []
    for i in range(len(dataframe)):
        degrees = dataframe[column][i]*pie_size
        sine.append(math.sin(degrees))
        cosine.append(math.cos(degrees))
    return sine, cosine
```

```python
time_features = ['weekday', 'month', 'day_of_year', 'day_of_month']
for feature in time_features:
    sine, cosine = sine_cosine( Final_Data, feature )
    Final_Data[feature+'_sin'] = sine
    Final_Data[feature+'_cos'] = cosine
```

```python
#create lists of load trend, weather, and time features for the forecasting day, _fd means forecasting day
columns_fd = ['1prev_week_load', '2prev_week_load', '3prev_week_load', '4prev_week_load', 'temp_Madrid',
                   'pressure_Madrid','humidity_Madrid','wind_speed_Madrid','temp_Seville','pressure_Seville',
                   'humidity_Seville','wind_speed_Seville', 'temp_Barcelona','pressure_Barcelona','humidity_Barcelona',
                   'wind_speed_Barcelona','temp_Valencia','pressure_Valencia','humidity_Valencia', 'wind_speed_Valencia', 
                   'temp_Bilbao','pressure_Bilbao','humidity_Bilbao', 'wind_speed_Bilbao', 
                   'weekday','month', 'day_of_year', 'day_of_month', 'weekdays_weekends', 'weekday_sin', 'weekday_cos',
                   'month_sin', 'month_cos', 'day_of_year_sin', 'day_of_year_cos', 'day_of_month_sin', 'day_of_month_cos']

for feature in columns_fd:
    fd_feature = [] 
    for i in range(0,len(Final_Data)):
        if i > len(Final_Data)-24*2-1 :
            fd_feature.append(0)
        else:
            fd_feature.append(Final_Data.loc[i+24*2, feature])
            
        #print(len(Final_Data[feature]))
    Final_Data[feature + '_fd'] = fd_feature
```

```python
#add noise in weather data, _n means noise
# call numpy.random.normal(m, s, shape) , m mean, s standard deviation, and  the shape of the original signal (original.shape)

collumns = ['temp_Madrid_fd', 'pressure_Madrid_fd', 'humidity_Madrid_fd', 'wind_speed_Madrid_fd', 
            'temp_Seville_fd','pressure_Seville_fd', 'humidity_Seville_fd','wind_speed_Seville_fd', 
            'temp_Barcelona_fd','pressure_Barcelona_fd','humidity_Barcelona_fd','wind_speed_Barcelona_fd', 
            'temp_Valencia_fd','pressure_Valencia_fd','humidity_Valencia_fd', 'wind_speed_Valencia_fd', 
            'temp_Bilbao_fd','pressure_Bilbao_fd','humidity_Bilbao_fd', 'wind_speed_Bilbao_fd' ]

for weather_feature in collumns:
    print(weather_feature)
    ## represent every column asarray
    weather = np.asarray(Final_Data[weather_feature])
    # find noise and add to original signal
    noise = np.random.normal(0, 1, weather.shape)
    print (noise)
    new_signal = weather + noise
    #add noisy column to dataframe 
    Final_Data[weather_feature+ '_n'] = new_signal
    
    
#drop the weather columns without noise
Final_Data = Final_Data.drop(['temp_Madrid', 'pressure_Madrid', 'humidity_Madrid', 'wind_speed_Madrid', 
            'temp_Seville','pressure_Seville', 'humidity_Seville','wind_speed_Seville', 
            'temp_Barcelona','pressure_Barcelona','humidity_Barcelona','wind_speed_Barcelona', 
            'temp_Valencia','pressure_Valencia','humidity_Valencia', 'wind_speed_Valencia', 
            'temp_Bilbao','pressure_Bilbao','humidity_Bilbao', 'wind_speed_Bilbao' ,
            'temp_Madrid_fd', 'pressure_Madrid_fd', 'humidity_Madrid_fd', 'wind_speed_Madrid_fd', 
            'temp_Seville_fd','pressure_Seville_fd', 'humidity_Seville_fd','wind_speed_Seville_fd', 
            'temp_Barcelona_fd','pressure_Barcelona_fd','humidity_Barcelona_fd','wind_speed_Barcelona_fd', 
            'temp_Valencia_fd','pressure_Valencia_fd','humidity_Valencia_fd', 'wind_speed_Valencia_fd', 
            'temp_Bilbao_fd','pressure_Bilbao_fd','humidity_Bilbao_fd', 'wind_speed_Bilbao_fd',
            'total load forecast', 'weekday', 'month','day_of_year', 'day_of_month' ,
            'weekdays_weekends', '1prev_week_load','2prev_week_load', '3prev_week_load', '4prev_week_load', 
            'weekday_sin','weekday_cos', 'month_sin', 'month_cos', 'day_of_year_sin',
            'day_of_year_cos', 'day_of_month_sin', 'day_of_month_cos',], axis = 1)
```

```python
#normalize data dividing by maximum value each column
data_columns = Final_Data.columns
data_col =data_columns[1:]
max_col = Final_Data.max()
i=1
for col in data_col:
    Final_Data[col] = Final_Data[col]/max_col[i]
    i+=1
```
